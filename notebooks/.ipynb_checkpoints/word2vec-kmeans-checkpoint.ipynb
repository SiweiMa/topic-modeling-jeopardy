## Import data

```python
with open('/content/drive/MyDrive/USF/spring2/msds626/finalproject/jeopardy.json') as f:
    data_json = json.load(f)
    
data = pd.DataFrame(data_json)
```

```python
category = data['category'].value_counts().index.to_list()
```

```python
print(f'The number of total records is {len(data)} and the number of unique categories is {len(category)}')
```

## Process the category

```python
import re
import string

nlp = spacy.load("en_core_web_sm")
nlp.vocab["category"].is_stop = True

def process(text):
    text = text.lower()
    text = re.sub('[' + string.punctuation + '0-9\\r\\t\\n]', ' ', text)
    text = " ".join([token.lemma_ for token in nlp(text) if token.is_stop is False]) #and (token.pos_ is 'PROPN' or token.pos_ is 'NOUN')
    words_list = text.split(' ')
    words_list = [w for w in words_list if len(w) > 2] # ignore a, an, to, at, be, ...
    processed_text = " ".join(words_list)

    return processed_text
```

```python
from tqdm import tqdm

processed_category = []
for c in tqdm(category):
    processed_category.append(process(c))
    
```

## Create new column with processed category

```python
mapping_category = {}
for i in range(len(category)):
    mapping_category[category[i]] = processed_category[i]
```

```python
with open('mapping_category.txt', 'w') as f:
    for i in mapping_category.items():
        f.write(str(i)+'\n')
```

```python
data['processed_category'] = data['category'].map(mapping_category)
```

```python
data_processed = data[data['processed_category']!='']
```

```python
print(f'The number of total original records is {len(data)} and the number of total processed records  is {len(data_processed)}')
```

## Convert unique category to word vector

```python
from gensim.models import KeyedVectors
from gensim.models import Word2Vec
```

```python
# model = KeyedVectors.load_word2vec_format('glove/glove.6B.300d.w2vformat.txt',binary=False, encoding='utf-8')
model = KeyedVectors.load_word2vec_format('/content/drive/MyDrive/USF/spring2/msds626/finalproject/GoogleNews-vectors-negative300.bin', binary=True)
```

```python
unique_processed_category = data_processed['processed_category'].unique()
processed_category_vector = []

for p in tqdm(unique_processed_category):
    words = p.split()
    vector = np.zeros(300)
    for w in words:
        try:
            vector += model.get_vector(w)
        except:
            break
    vector = vector/len(words)
    processed_category_vector.append(vector)
        
```

```python
from sklearn.preprocessing import normalize

X = np.array(processed_category_vector)
X_normalize = normalize(X)
```

```python
print(f'The number of unique processed categories is {len(X_normalize)}')
```

```python
mapping_vector = {}
for i in range(len(unique_processed_category)):
    mapping_vector[unique_processed_category[i]] = X_normalize[i]
```

## Cluster word vectors

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist

# credit: https://www.geeksforgeeks.org/elbow-method-for-optimal-value-of-k-in-kmeans/
# k means determine k
def elbow(X, n_cluster):
    distortions = []
    K = range(1,n_cluster+1)
    for k in tqdm(K):
        kmeanModel = KMeans(n_clusters=k).fit(X)
        distortions.append(sum(np.min(cdist(X, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / X.shape[0])

    # Plot the elbow
    plt.plot(K, distortions, 'bx-')
    plt.xlabel('k')
    plt.ylabel('Distortion')
    plt.title('The Elbow Method showing the optimal k')
    plt.show()
    return K, distortions
```

```python
n_cluster = 20
K, distortions = elbow(X_normalize, n_cluster)
```

```python
from kneed import DataGenerator, KneeLocator
kneedle = KneeLocator(K, distortions, S=0.1, curve="concave", direction="decreasing")
print(round(kneedle.knee, 3))
```

```python
n_cluster = 15

kmeanModel = KMeans(n_clusters=n_cluster).fit(X_normalize)
labels = kmeanModel.labels_

labels
```

## Assign topics

```python
"""
combine the words of categories in the same cluster into a single string
"""
categories_text = []
for i in np.unique(labels):
    words_in_category = []
    categories = np.array(list(unique_processed_category))[labels == i]
    for c in categories:
        for e in c.split():
            words_in_category.append(e)
    categories_text.append(' '.join(words_in_category))
```

```python
def get_word_count(scripts):
    """
    summarize the word count of the input string
    output the data frame with index as word count and columns as vocabulary
    """
    vectorizer = CountVectorizer()
    X = vectorizer.fit_transform(scripts)
    word_count_matrix = pd.DataFrame(data=X.toarray(), columns=vectorizer.get_feature_names())

    return word_count_matrix
```

```python
label_name = ['American', 'History', 'Celebrity', 
              'Music', 'Book', 'Miscellaneous', 
              'Miscellaneous', 'Sport', 'Society/family',
              'Food',  'Word', 'Miscellaneous',
              'Politics', 'Animal', 'Movie']
```

```python
top_n = 20

fig, axes = plt.subplots(5, 3, figsize=(30, 40), sharex=True)
axes = axes.flatten()
y_pos = range(top_n, 0, -1)
for i in np.unique(labels): #range(n_cluster):
    word_count_matrix_sorted = get_word_count([categories_text[i]]).sort_values(by=0, axis=1, ascending=False)
    axes[i].barh(y_pos, word_count_matrix_sorted.values[0][:top_n])
    axes[i].set_yticks(y_pos)
    axes[i].set_yticklabels(word_count_matrix_sorted.columns[:top_n])
    axes[i].set_title(f'Topic {label_name[i]}')
```

```python
mapping_label = {}
for i in range(len(unique_processed_category)):
    mapping_label[unique_processed_category[i]] = labels[i]
```

```python
mapping_label_name = {}
for i in np.unique(labels):
    mapping_label_name[i] = label_name[i]
```

```python
data_processed['label'] = data_processed['processed_category'].map(mapping_label)
data_processed['topic'] = data_processed['label'].map(mapping_label_name)
```
